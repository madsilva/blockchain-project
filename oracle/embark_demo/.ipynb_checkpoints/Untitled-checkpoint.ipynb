```python
import requests
from bs4 import BeautifulSoup
```

```python
url="http://alisondb.legislature.state.al.us/alison/codeofalabama/constitution/1901/toc.htm"
```

```python
url2="http://alisondb.legislature.state.al.us/alison/codeofalabama/constitution/1901/"
```

```python
r = requests.get(url)
sufs=[]
soup = BeautifulSoup(r.text, 'html.parser')
for link in soup.find_all('a'):
    sufs.append(link.get('href'))
```

```python
text=''
for suf in sufs:
    #r = requests.get(url2+suf)
    r=requests.get("http://alisondb.legislature.state.al.us/alison/codeofalabama/constitution/1901/{suf}}")
    s = BeautifulSoup(r.text, 'html.parser')
    text+=(s.get_text())
```
